```python
Y_feature = df['motor_UPDRS']
X_features = df.drop(columns = ['motor_UPDRS'])
X_features.drop(columns =  ['Jitter(Abs)', 'Jitter:RAP', 'Jitter:PPQ5', 'Jitter:DDP', 'Shimmer', 'Shimmer(dB)', 'Shimmer:APQ3', 
                           'Shimmer:APQ5', 'Shimmer:DDA'], inplace=True)
```

```python
#train-test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_features,Y_feature, test_size=0.2, random_state = 42)
```

```python
print(f'Training Dataset has {len(X_train)} voice measurements')
print(f'Test Dataset has {len(X_test)} voice measurements')
```

```python
from sklearn.preprocessing import MinMaxScaler
mm_scaler = MinMaxScaler()
X_train = pd.DataFrame(mm_scaler.fit_transform(X_train), columns = X_train.columns) 
X_test = pd.DataFrame(mm_scaler.fit_transform(X_test), columns = X_test.columns)


X_train['x1'] = np.ones((len(X_train),1))
X_test['x1'] = np.ones((len(X_test),1))
```

```python
X_train
```

```python
class LinearRegression:
    def __init__(self, iterations = 1000, learning_rate = 0.01, beta = 0.9, momentum=True):
        self.iterations = iterations
        self.learning_rate = learning_rate
        self.weights = None
        self.momentum = momentum
        self.costs = np.zeros(iterations)
        self.beta = beta
        
    def fit(self, X, y):
        num_examples, num_features = X.shape
        self.weights = np.ones(num_features) 
        vdW = np.zeros(num_features)
        
        for iteration in range(self.iterations):
            weighted_sum = np.dot(X_train, self.weights)
            loss = weighted_sum - y_train
            cost = np.sum(np.square(loss))/(2*num_examples)
            
            #Gradient w.r.t to parameters
            dW = np.dot(X_train.T, loss)/num_examples
            
            if(self.momentum == True):
                vdW = (self.beta*vdW) + (1-self.beta)*dW
                self.weights -= self.learning_rate*vdW
            else:
                self.weights -= self.learning_rate*dW

            self.costs[iteration] = cost
            
#             if(abs(self.costs[iteration] - self.costs[iteration-1]) < 0.05):
#                 break
    
            
    def predict(self,X_test):
        return np.dot(X_test, self.weights)    
    
    def mean_squared_error(self, predictions, true_values):
        return np.sum(np.square(predictions-true_values))/len(true_values)
    
```

```python
batch_model = LinearRegression(iterations = 1000, learning_rate = 0.01, momentum=False)
batch_model.fit(X_train, y_train)

momentum_model = LinearRegression(iterations = 1000, learning_rate = 0.01, momentum=True)
momentum_model.fit(X_train, y_train)
```

```python
y_pred = batch_model.predict(X_test)
print(batch_model.mean_squared_error(y_pred, y_test))
```

# Batch Gradient Descent vs Batch Gradient Descent with Momentum

```python
plt.plot(list(range(batch_model.iterations)), batch_model.costs, 'red')
plt.plot(list(range(momentum_model.iterations)), momentum_model.costs, 'blue' )
plt.title("Batch GD vs Momentum GD")
plt.xlabel("# of iterations")
plt.ylabel("cost")
plt.legend(['Batch GD', 'Momentum GD'])
```

```python
lowest_value = 1000000000000000000
for alpha in [10**-3, 10**-2, 10**-1, 10**0]:
    print(f"alpha value: {alpha}")
    for iteration in [10, 100,1000, 10000]:
        print(f"Number of Iterations = {iteration}")
        for beta in [0.8,0.9]:
            print(f"alpha = {alpha} iterations = {iteration} and beta = {beta }")
            model = LinearRegression(iteration, learning_rate = alpha, beta = beta, momentum=True)
            model.fit(X_train, y_train)
            y_pred = model.predict(X_test)
            print("\tPerformance with learning rate as {} is {}".format(alpha, model.mean_squared_error(y_pred, y_test)))
            if(model.mean_squared_error(y_pred, y_test) < lowest_value):
                lowest_value = model.mean_squared_error(y_pred, y_test)
                
```

```python
plt.plot(list(range(model.iterations)), model.costs, 'red')
plt.title("Cost function without Momentum Term ")
plt.xlabel("# of iterations")
plt.ylabel("cost")
```
